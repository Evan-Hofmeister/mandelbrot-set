```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define function ot create find Mandelbrot or Julia set

The mandelbrot set is defined by the set of complex numbers which remain bounded in absolute value. The sequence  is defined by:
$$z_0 = 0$$
$$z_{n+1} = z^{2}_{n} + c$$

```python
# Function takes in intial value of c and n number of max iterations
def mandelbrot_set(z, c, n):
    
    for iteration in range(n):
        z=(z**2)+c
        
        # Rough threshold dividing value where zed diverges
        if abs(z) > 2: 
            break
    
    return iteration
```

### Define function to populate atlas (showing convergence or divergence)

```python
def populate_atlas(set_choice, set_param, xvalues, yvalues):
    
    # define size of containers
    xlen = len(xvalues)
    ylen = len(yvalues)
    
    # Container to hold number of iterations for each set in complex plane
    atlas = np.zeros((xlen,ylen))

    # Use Madelbrot function to check for convergence (part of mandelbrot set) or divergence
    for i_x in range(xlen):
        for i_y in range(ylen):

            cz_x = xvalues[i_x]
            cz_y = yvalues[i_y]
            
            # For mandelbrot, we start z at zero, and c is the point being tested
            if set_choice == 'mandelbrot':
                z = set_param
                c = complex(cz_x, cz_y) 
            
            # for Julia, we specify c and hold it constant, and z is the point being tested 
            elif set_choice == 'julia':
                c = set_param
                z = complex(cz_x, cz_y) 
                
            else:
                break

            
            atlas[i_x,i_y] = mandelbrot_set(z, c,120)

    # plot
    plt.figure(figsize=(10, 10), dpi=80)
    plt.imshow(atlas.T, interpolation="nearest", cmap="jet")
    plt.show()
```

### Plot Resuts

Note the colors not only represent convergence or divergence but also speed of divergence


```python
# Entire complex plane
x = np.linspace(-2.25, 0.75, 1000)
y = np.linspace(-1.5, 1.5, 1000)

set_choice = 'mandelbrot'
set_param = complex(0,0)

populate_atlas(set_choice, set_param, x, y)
```

```python
# A closer look
xvalues = np.linspace(-0.25, -0.15, 1000)
yvalues = np.linspace(-0.75, -0.65, 1000)

set_choice = 'mandelbrot'
set_param = complex(0,0)

populate_atlas(set_choice, set_param, xvalues, yvalues)
```

### Julia set

Note that if c is within the mandelbrot set, you will see a distinct singular blob, but as you move c away from the mandelbrot set the more detached the julia set appears 

```python
xvalues = np.linspace(-1.75, 1.75, 1000)
yvalues = np.linspace(-1.75, 1.75, 1000)

set_choice = 'julia'
set_param = complex(-0.166517,-0.662029)
populate_atlas(set_choice, set_param, xvalues, yvalues)

set_param = complex(-0.18,-0.66)
populate_atlas(set_choice, set_param, xvalues, yvalues)

set_param = complex(-0.22,-0.66)
populate_atlas(set_choice, set_param, xvalues, yvalues)

set_param = complex(-0.26,-0.66)
populate_atlas(set_choice, set_param, xvalues, yvalues)

set_param = complex(-0.35,-0.76)
populate_atlas(set_choice, set_param, xvalues, yvalues)
```
